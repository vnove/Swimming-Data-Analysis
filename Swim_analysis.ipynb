**Left Stroke Analysis**

```python
data = pd.read_csv('Lstrokeanalysis.csv') 
```

```python
correlation_matrix = data.corr(method = 'kendall')
print(correlation_matrix)
```

```python
sns.scatterplot(x="Propolsive", y="PaD", data=data)
```

```python
# First, select the variables you want to plot against 'Propolsive'
selected_vars = ["PaD", "vertical_pressure", "realtive_displ_z", "realtive_velo_x", "PPU_X", "realtive_displ_y", "linear_acceleration_z"]

# Melt the DataFrame to long format
melted = data.melt(id_vars=["Propolsive"], value_vars=selected_vars, var_name="Variable", value_name="Value")

# Create a FacetGrid
g = sns.FacetGrid(melted, col="Variable", col_wrap=3, sharex=False, sharey=False)
g.map(sns.scatterplot, "Propolsive", "Value")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = data[[
    "PaD", 
    "vertical_pressure", 
    "realtive_displ_z", 
    "realtive_velo_x", 
    "PPU_X", 
    "realtive_displ_y", 
    "linear_acceleration_z"
]]

y = data["Propolsive"]

model = LinearRegression ()
model.fit(X, y)

impact = pd.DataFrame({
    "Varaible": X.columns,
    "Coefficient": model.coef_
}).sort_values(by="Coefficient", ascending=False)

print(impact)
```

```python
frame_time = 0.01
data["impulse_contribution"] = data["Propolsive"] * frame_time

total_impulse = data["impulse_contribution"].sum()
print(f"Total Propulsive Time: {total_impulse:.6f}  N·s")

data["speed"] = np.sqrt(
    data["realtive_velo_x"]**2 +
    data["realtive_velo_y"]**2 +
    data["realtive_velo_z"]**2
)

avg_speed = data["speed"].mean()
print(f"Average Speed: {avg_speed:.3f} m/s")
```

```python
predictors = [
    "PaD", 
    "vertical_pressure", 
    "realtive_displ_z", 
    "realtive_velo_x", 
    "PPU_X", 
    "realtive_displ_y", 
    "linear_acceleration_z"
]
X = data[predictors]
y = data["speed"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())

results = model.summary2().tables[1]
print("Recommended variables (p < 0.05): ")
print(results["P>|t|"] < 0.05)
```
